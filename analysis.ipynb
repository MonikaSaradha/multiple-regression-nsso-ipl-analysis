# Multiple Regression & Diagnostics (NSSO & IPL)

This notebook applies regression modeling, diagnostics, and ridge regression using NSSO and IPL datasets.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import r2_score
from scipy.stats import shapiro
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan
```

## NSSO Dataset

```
data = pd.read_excel("ASSG1.xlsx")

data['tot_con'] = (
    data['wheattotal_q'] + data['cerealtot_q'] + data['moong_q'] +
    data['pulsestot_q'] + data['milk_q'] + data['onion_q'] + data['potato_q']
)

features = ['wheattotal_q','cerealtot_q','moong_q','pulsestot_q','milk_q','onion_q','potato_q']
X = data[features]
y = data['tot_con']
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

### Linear Regression

```
model = LinearRegression()
model.fit(X_train, y_train)

pred = model.predict(X_test)

print("R-squared:", r2_score(y_test, pred))
```

### MAPE

```
def MAPE(actual, predicted):
    return np.mean(np.abs((actual - predicted) / actual)) * 100

print("MAPE:", MAPE(y_test, pred))
```

### Diagnostics

```
residuals = y_test - pred

plt.scatter(pred, residuals)
plt.title("Residuals vs Predicted")
plt.show()
```

```
print("Shapiro p-value:", shapiro(residuals).pvalue)
```

```
X_const = sm.add_constant(X_test)
bp_test = het_breuschpagan(residuals, X_const)
print("Breusch-Pagan p-value:", bp_test[1])
```

### Ridge Regression

```
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)

ridge_pred = ridge.predict(X_test)
print("Ridge R-squared:", r2_score(y_test, ridge_pred))
```

## IPL Dataset

```
ipl = pd.read_excel("IPL salary 2008_2022.xlsx")

ipl = ipl[['Runs', 'Wkts', 'Final Price']].dropna()

X_ipl = ipl[['Runs', 'Wkts']]
y_ipl = ipl['Final Price']
```

```
model_ipl = LinearRegression()
model_ipl.fit(X_ipl, y_ipl)

pred_ipl = model_ipl.predict(X_ipl)

print("R-squared (IPL):", r2_score(y_ipl, pred_ipl))
```

```
residuals_ipl = y_ipl - pred_ipl

plt.scatter(pred_ipl, residuals_ipl)
plt.title("Residual Plot (IPL)")
plt.show()
```

```
plt.hist(residuals_ipl)
plt.title("Residual Distribution")
plt.show()
```

## Key Insights

- High R-squared indicates possible overfitting in NSSO model
- Residual tests suggest violations of assumptions
- Ridge regression improves stability
- IPL model shows strong link between performance and salary
